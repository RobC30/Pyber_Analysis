```python
%matplotlib inline
%matplotlib notebook

# Import dependencies.
import matplotlib.pyplot as plt
import statistics
import numpy as np
import pandas as pd
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color='b', linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
#Add a grid
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, marker='D',color='g', linewidth=2, label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_ylim(0,45)
ax.grid()
ax.legend()
```

```python
# Create the BAR plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the BARh plot (inverted y_axis for better visualisation)
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis,color='cyan',label='Chicago')
ax.invert_yaxis()
ax.set_title('PyBer by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.legend()
```

```python
# Create a scatter plot
plt.scatter(y_axis, x_axis, c='red',label='Chicago',s=y_axis)
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i*10 for i  in y_axis], color='skyblue', edgecolors='black', linewidth=2, alpha=.2 )
ax.invert_yaxis()
ax.set_xlim(0,50)
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.subplots(figsize=(8, 8))
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# create pie chart using OOA
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis, startangle=90, counterclock=False)

plt.show()
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=7)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(10,10))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
